# Data Preprocessing
Step 1 : Importing the neccessary libraries.

```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import RidgeCV, LassoCV, Ridge, Lasso
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import RandomizedSearchCV
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
from sklearn.datasets import fetch_california_housing
from sklearn import datasets
```

Step 2 : Import the dataset

```
california_data=datasets.fetch_california_housing()
data= pd.DataFrame(california_data.data, columns= california_data.feature_names)
data['target'] = pd.Series(california_data.target)
```

Step 3 : Check out the missing values


> We do not have any missing values



```
data.isna().sum()
```

Step 4 : Deal with categorical values

>Since, machine learning models are based on Mathematical equations and you can intuitively understand that it would cause some problem if we can keep the Categorical data in the equations because we would only want numbers in the equations. 


> We do not have any categorical values





```
X = data.iloc[: , : -1].values
X
```

#Feature Selection


    Method 1 : Pearson Correlation



```
plt.figure(figsize=(12,10))
cor = data.corr()
sns.heatmap(cor, annot=True, cmap=plt.cm.Reds)
plt.show()
```

The correlation matrix shows that MedInc is highly correlated to the target with a correlation coefficient of 0.69

AveBedrms and AveRooms are highly corrleated independent variables which would affect the model perfomance. We therefore keep AveRooms since it has a higher correlation(0.15) with the target as compared to AveBedrms abs(-0.047)

We also drop Longitude because it is highly correlated to latitude and is less correlated to the target.

**Method 2: Embedded Method using Lasso Regularisation**

```
X = data.drop("target",1)   
y = data["target"] 
reg = LassoCV()
reg.fit(X, y)
print("Best alpha using built-in LassoCV: %f" % reg.alpha_)
print("Best score using built-in LassoCV: %f" %reg.score(X,y))
coef = pd.Series(reg.coef_, index = X.columns)
```

```
imp_coef = coef.sort_values()
import matplotlib
matplotlib.rcParams['figure.figsize'] = (8.0, 10.0)
imp_coef.plot(kind = "barh")
plt.title("Feature importance using Lasso Model")
```

The Lasso Regularization picked MedInc, HouseAge, AveRooms, AveOccup, Lattitude and Longitude and dropped AveBedrms and Population

Based on Correlation coefficient and Lasso regularisation, we drop AveBedrms and Population

**Drop features based on high correlation and feature importance**

```
data.drop(['AveBedrms','Population','Longitude'],axis=1, inplace=True)
```

Assign X and Y values
X -> Independent Variables
Y -> Target Variable

```
X=data.loc[:,['MedInc','HouseAge','AveRooms','Latitude']]
y=data.loc[:,['target']].values
```

**Standardize the values to make sure there is internal consistency**

```
X=StandardScaler().fit_transform(X)
```

**Apply Principal Component Analysis**

```
Pca=PCA(.90)

pcs=Pca.fit_transform(X)

pcs.shape
```

```
Pca.explained_variance_ratio_
```

**You will need 4 Principal Components to get a variance greater than 90**

```
vis = sns.pairplot(data)
```

From the histograms, we can tell that MedInc,AveOccup and AveRooms are positively skewed.

The pair plot shows that there is a positive correlation between MedInc and Target i.e. as Median Income increases so does price

**Linear Regression**

Start by splitting the dataset

```
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

Fit the model


```
regressor = LinearRegression()  

regressor.fit(X_train, y_train)
```

Retrieve the intercept and the slope

```
print(regressor.intercept_)
print(regressor.coef_)
```

#Predict California House Prices

```
y_pred = regressor.predict(X_test)
```

```
df = pd.DataFrame({'Actual': y_test.flatten(), 'Predicted': y_pred.flatten()})
df
```

**Check the regression score**

```
regressor.score(X_train,y_train)
```

**The linear regression model accurately predicts with an accuracy of 52.4%**

#Hyperparameter Tuning
1. Ridge Regression
2. Grid Search

```
model = Ridge()
ridge_params = {'alpha':[0.02, 0.024, 0.025, 0.026, 0.03],'max_iter':[1000,1100,2000],'tol':[0.001,0.0001,0.000001]}
#param_grid = dict(alpha = lasso_params)
grid = GridSearchCV(model,param_grid=ridge_params,cv=2,n_jobs=-1)
#grid.fit(X,y)
results = grid.fit(X_train,y_train)
#print(results.best_score_)
#print("Best Params are ", results.best_params_)
results.score(X_test,y_test)
```

**Our score is now 49%**

**Randomized Search**

```
random = RandomizedSearchCV(model,param_distributions=ridge_params,cv=3,n_jobs=-1)
results = random.fit(X_train,y_train)
print(results.best_score_)
print("Best Params are ", results.best_params_)
```
